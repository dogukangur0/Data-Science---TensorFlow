```python
dataFrame = nintyninepercentDataFrame
```

```python
dataFrame = dataFrame[dataFrame.year != 1970]
```

```python
dataFrame
```

```python
dataFrame = dataFrame.drop("transmission",axis=1)
```

```python
dataFrame.head()
```

```python
# x ve y bul
```

```python
y = dataFrame["price"].values
x = dataFrame.drop("price",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)

# y verisini scaler etmeye gerek yok
```

```python
from tensorflow.keras.models import Sequential  # create model
from tensorflow.keras.layers import Dense, Dropout       # create layer (katman)
```

```python
model = Sequential()

model.add(Dense(128,activation="relu"))
model.add(Dense(64,activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(32,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
# validasyon verisini tf yardımıyla modelin içerisine verebiliriz.
# eğitilirken valide edilebilir.
# daha önce manuel yapılan x_test,y_test ile kıyaslama işlemlerini daha kolay hale getiriyor. 
model.fit(x=x_train, y=y_train, validation_data=(x_test,y_test), batch_size=250, epochs=300)
```

```python
loss = pd.DataFrame(model.history.history)
loss.head()
```

```python
loss.plot()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
predictArray = model.predict(x_test)
predictArray
```

```python
mean_squared_error(y_test,predictArray)
```
