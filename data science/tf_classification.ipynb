```python
y = dataFrame["Type"].values
x = dataFrame.drop("Type",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.fit_transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping

# EarlyStopping ve Dropout'u overfitting problemlerini çözmek için kullanırız.
```

```python
x_train.shape
```

```python
model = Sequential()

# shape içerisinde kaç tane column varsa units değerini ona eşit yapmamız önerilir.
# diğer son layer = 1, ilk layer = 30 , aradaki layerlar ise 1 ve 30 aralığında olması önerilir.
model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))

# sigmoid 0 ve 1 arasında değer verir.
# sınıflandırma problemlerinde output layer'a sigmoid activation kullanarak bunu elde edebiliriz.

model.compile(loss="binary_crossentropy",optimizer="adam")
# sınıflandırma olduğu için loss -> binary 0-1
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1)
```

```python
loss = pd.DataFrame(model.history.history)
loss
```

```python
loss.plot()
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
earlyStopping = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
loss = pd.DataFrame(model.history.history)
```

```python
loss.plot()
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1,activation="sigmoid"))

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
loss = pd.DataFrame(model.history.history)
loss.plot()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
predictions = model.predict(x_test)
predicted_classes = (predictions > 0.5).astype("int32")
```

```python
print(classification_report(y_test,predicted_classes))
```

```python
print(confusion_matrix(y_test,predictions))
```
