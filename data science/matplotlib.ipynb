```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ageData = [10,20,30,40,50,60]
weightData = [40,70,80,74,65,68]
```

```python
ageDataNp = np.array(ageData)
weightDataNp = np.array(weightData)
```

```python
plt.plot(ageDataNp,weightDataNp,"r")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("AGE-WEIGHT")
```

```python
plt.plot(ageDataNp,weightDataNp,"r--")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("AGE-WEIGHT")
```

```python
plt.plot(ageDataNp,weightDataNp,"r*-")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("AGE-WEIGHT")
```

```python
plt.plot(ageDataNp,weightDataNp,"r*")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("AGE-WEIGHT")
```

```python
# birden fazla grafik oluşturma

plt.subplot(1,2,1) # kaç sıra olacak, sırada kaç tane grafik olacak, kaçıncı grafiği çiziyorsun
plt.plot(ageDataNp,weightDataNp,"b")
plt.subplot(1,2,2)
plt.plot(weightDataNp,ageDataNp,"y")
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(ageDataNp,weightDataNp,"g")
figureAxes.set_xlabel("Age")
figureAxes.set_ylabel("Weight")
figureAxes.set_title("Age-Weight")
```

```python
figureTwo = plt.figure()

axisOne = figureTwo.add_axes([0.1,0.1,0.8,0.8])
axisTwo = figureTwo.add_axes([0.3,0.2,0.4,0.4])

axisOne.plot(ageDataNp,weightDataNp,"g")
axisOne.set_xlabel("Age")
axisOne.set_ylabel("Weight")
axisOne.set_title("Age - Weight")

axisOne.plot(ageDataNp,weightDataNp,"r")
axisOne.set_xlabel("Age")
axisOne.set_ylabel("Weight")
axisOne.set_title("Age - Weight")
```

```python
# subplots

(myFigure,myAxis) = plt.subplots(nrows=1,ncols=2)

for axis in myAxis:
    axis.plot(ageDataNp,weightDataNp,"g")
    axis.set_xlabel("Age")

plt.tight_layout()
```

```python
myFigure = plt.figure(figsize=(6,6))  # bu içerisine figsize değeri verirsek, oluşturduğumuz axis üzerine yazar.
myAxis = myFigure.add_axes([0.1,0.1,0.9,0.9])
myAxis.plot(ageDataNp,ageDataNp**2/2,"purple")
myAxis.plot(ageDataNp,ageDataNp*70,"gray")
```

```python
myFigure.savefig("myfig.png",dpi=200)
```
